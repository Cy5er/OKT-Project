```python
# Import dependencies
import pandas as pd
import numpy as np
import ast
from bs4 import BeautifulSoup
```

```python
# Read the original data
dataset = pd.read_pickle('data'+ '/dataset.pkl')
dataset.head()
```

```python
# Read the code_samples
df_code = pd.read_csv('./FalconCodeCSV/falconcode_v1_code_samples.csv')
df_code.head()
```

```python
# Read the problems
df_prompt = pd.read_csv('./FalconCodeCSV/falconcode_v1_table_problems.csv')
df_prompt.head()
```

```python
# Read the runs
df_scores = pd.read_csv('./FalconCodeCSV/falconcode_v1_table_runs.csv')
df_scores.head()
```

```python
# Drop unncessary columns
df_scores = df_scores.drop('id', axis = 1)
df_scores.head()
```

```python
# Rename columns to match original data
df_scores.rename(columns={'code_hash':'codestate'}, inplace= True)
df_scores.head()
```

```python
# Rename columns to match original data
df_code.rename(columns={'hash':'codestate'}, inplace= True)
df_code.head()
```

```python
# Merge the score and code dataframes and get rid of uncessary columns
merged_df = pd.merge(df_code,df_scores,on='codestate')
merged_df = merged_df.drop('redacted', axis = 1)
merged_df.head()
```

```python
# Convert timestamp into datetime, sort and delete all intermediate submissons
merged_df['timestamp'] = pd.to_datetime(merged_df['timestamp'])
merged_df.sort_values(by = ['student_id','problem_id','timestamp'], inplace= True, ascending=[False,False,True])
merged_df = merged_df.drop_duplicates(subset=['student_id','problem_id'], keep= 'last')
merged_df.head(10)
```

```python
df_prompt.rename(columns={'id':'problem_id'}, inplace= True)
df_prompt.head()
```

```python
specific_row = df_prompt[df_prompt['problem_id'] == 'Lists3_lowvolume']
specific_row
```

```python
#merged_df = pd.merge(merged_df,df_prompt[['problem_id','max_score']], on = 'problem_id')
df_prompt = df_prompt.drop_duplicates(subset='problem_id')
# Update merged_df with unique 'problem_id' values from df_prompt
merged_df['max_score'] = merged_df['problem_id'].map(df_prompt.set_index('problem_id')['max_score'])
merged_df.head(20)
```

```python
merged_df['score_x'] = 1.00 * merged_df['score'] / merged_df['max_score']

# Clip the values of 'score_x' between 0 and 1
merged_df['score_x'] = merged_df['score_x'].clip(lower=0, upper=1)
```

```python
merged_df = merged_df.drop(['score', 'max_score'], axis=1)
merged_df.head()
```

```python
merged_df['score_y'] = np.select([merged_df['score_x'] == 1, merged_df['score_x'] == 0], [2, 0], default=1)
merged_df.head()
```

```python
merged_df['Subject_id'] = merged_df.index
merged_df = merged_df.sort_values(by='Subject_id')
merged_df.insert(0, 'Subject_id', merged_df.pop('Subject_id'))
merged_df = merged_df.drop('timestamp', axis = 1)
merged_df.head()
```

```python
merged_df.rename(columns={'course_id':'AssignmentID'}, inplace= True)
merged_df.head()
```

```python
merged_df['prompt'] = merged_df['problem_id'].map(df_prompt.set_index('problem_id')['prompt'])
```

```python
tree = ast.parse(test_code)

# Now 'tree' contains the abstract syntax tree representation of the source code
# You can explore and analyze the AST as needed

# For example, you can use ast.dump to visualize the AST
print("Abstract Syntax Tree:")
print(ast.dump(tree))
```

```python
def parse_source_code(code):
    try:
        tree = ast.parse(code)
        return ast.dump(tree)
    except SyntaxError:
        return None
```

```python
# Apply the parsing function to the 'source_code' column
merged_df['code_ast'] = merged_df['source_code'].apply(parse_source_code)

# Drop rows where AST parsing failed (ast_tree is None)
merged_df = merged_df.dropna(subset=['code_ast'])

merged_df.head()
```
