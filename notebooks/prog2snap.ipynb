```python
import pandas as pd
import numpy as np
```

```python
df_code = pd.read_csv('./FalconCodeCSV/falconcode_v1_code_samples.csv')
df_code.head()
```

```python
df_prompt = pd.read_csv('./FalconCodeCSV/falconcode_v1_table_problems.csv')
df_prompt.head()
```

```python
df_scores = pd.read_csv('./FalconCodeCSV/falconcode_v1_table_runs.csv')
df_scores.head()
```

```python
df_scores['timestamp'] = pd.to_datetime(df_scores['timestamp'])
df_scores = df_scores.sort_values(by=['timestamp'])
```

```python
df_scores.head()
```

```python
df_students = pd.read_csv('./FalconCodeCSV/falconcode_v1_table_students.csv')
df_students.head()
```

```python
merged_df = pd.merge(df_scores, df_students, left_on='student_id', right_on='id', how='inner')
merged_df = merged_df.drop(['id_y','course_id_y','stem_major','gender','ethnicity'], axis = 1)
merged_df
```

```python
merged_df = pd.merge(merged_df, df_prompt, left_on='problem_id', right_on='id', how='inner')
merged_df['ToolInstances'] = 'Python'
merged_df['EventType'] = 'Submission'
merged_df = merged_df.drop(["id",'course_id_x',"type","exam","assignment","function_return","testcase","input_str","input_cast","conditional","function_call","function_def","loop_counting","loop_until","loop_elements","loop_nested","stat_calculate","file_read","file_write","list","list_2d","dictionary","item_set","tuple"], axis = 1)
merged_df
```
